```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score , mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import load_diabetes 

data = load_diabetes()
```

```python
print(data.DESCR)
```

```python
X = data.data
y = data.target
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y , test_size = 0.2 , random_state=45)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train , y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
print("R2 score" , r2_score(y_test , y_pred))
print("RMSE" ,np.sqrt(mean_squared_error(y_test , y_pred)))
```

Using Ridge Regression

```python
from sklearn.linear_model import Ridge 
R = Ridge(alpha=0.0001)
```

```python
R.fit(X_train , y_train)
```

```python
y_pred1 = R.predict(X_test)
```

```python
print("R2 score" , r2_score(y_test , y_pred1))
print("RMSE" ,np.sqrt(mean_squared_error(y_test , y_pred1)))
```

```python
 m = 100
x1 = 5 * np.random.rand(m,1) - 2
x2 = 0.7 * x1** 2 - 2 * x1 + 3 + np.random.randn(m ,1)

plt.scatter(x1 , x2)
plt.show()
```

```python
from sklearn.pipeline import Pipeline 
from sklearn.preprocessing import PolynomialFeatures
```

```python
def get_preds_ridge(x1,x2,alpha):
    model = Pipeline([
        ('poly_feats' , PolynomialFeatures(degree=16)),
        ('ridge', Ridge(alpha=alpha))
    ])
    model.fit(x1,x2)
    return model.predict(x1)

alphas = [0 , 20, 200]
cs = ['r' , 'g' , 'b']

plt.figure(figsize=(10,6))
plt.plot(x1,x2, 'b+' , label='Darapoints')

for alpha , c in zip(alphas, cs):
    preds = get_preds_ridge(x1,x2,alpha)
    #plot 
    sort_idx = np.argsort(x1[:, 0])
    sorted_x1 = x1[sort_idx]
    sorted_preds = preds[sort_idx]

    plt.plot(sorted_x1, sorted_preds, c, label='Alpha: {}'.format(alpha))


plt.legend()
plt.title("Ridge Regression with Different Alphas")
plt.xlabel("x1")
plt.ylabel("x2 / Predictions")
plt.grid(True)
plt.show()    
```
